# Doing the same for companies df

```python
# Convert Create Date to datetime
companies_df['Create Date'] = pd.to_datetime(companies_df['Create Date'], errors='coerce')

# Create date-based fields
companies_df['Create_Year'] = companies_df['Create Date'].dt.year
companies_df['Create_Month'] = companies_df['Create Date'].dt.month
companies_df['Create_Quarter'] = companies_df['Create Date'].dt.quarter
companies_df['Create_YearMonth'] = companies_df['Create Date'].dt.to_period('M').astype(str)

# 1. Standardized Industry (simple title-case normalization)
companies_df['Industry_Standardized'] = companies_df['Industry'].str.title().str.strip()

# 2. Company Size Category (based on 'Total Agents')
# Use quantile binning for 'Number of Employees' into 5 categories
# This ensures roughly equal-sized groups

# Drop NaNs temporarily for binning
employee_data = companies_df[['Number of Employees']].dropna()

# Create quantile bins
companies_df['Company_Size_Category'] = pd.qcut(
    companies_df['Number of Employees'],
    q=5,
    labels=['Very Small', 'Small', 'Medium', 'Large', 'Very Large']
)



# 3. Revenue Category
def revenue_category(revenue):
    if pd.isnull(revenue):
        return 'Unknown'
    elif revenue < 1e6:
        return '<$1M'
    elif revenue < 1e7:
        return '$1M-$10M'
    elif revenue < 5e7:
        return '$10M-$50M'
    elif revenue < 1e8:
        return '$50M-$100M'
    else:
        return '>$100M'

companies_df['Revenue_Category'] = companies_df['Annual Revenue'].apply(revenue_category)

# 4. Region mapping (simplified)
def map_region(country):
    if pd.isnull(country):
        return 'Unknown'
    country = country.strip().lower()
    if country in ['united states', 'canada', 'mexico']:
        return 'North America'
    elif country in ['brazil', 'argentina', 'chile']:
        return 'South America'
    elif country in ['united kingdom', 'germany', 'france', 'netherlands']:
        return 'Europe'
    elif country in ['india', 'china', 'japan', 'singapore']:
        return 'Asia'
    elif country in ['australia', 'new zealand']:
        return 'Oceania'
    else:
        return 'Other'

companies_df['Region'] = companies_df['Country/Region'].apply(map_region)

# 5. Technology_Count from "Web Technologies"
companies_df['Technology_Count'] = companies_df['Web Technologies'].fillna('').apply(lambda x: len(x.split(';')) if x else 0)

# 6. Uses_[Technology] flags (example: uses Google Tag Manager)
companies_df['Uses_Google_Tag_Manager'] = companies_df['Web Technologies'].fillna('').str.contains('Google Tag Manager', case=False).astype(int)

# 7. Is_BPO flag
companies_df['Is_BPO'] = companies_df['BPO Program'].notna().astype(int)
```

# Working on choosing the features and building models

```python
# Cell 1 (corrected): Import libraries, load data, and drop all date-related columns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Load dataset
df = binary_df.copy()

# Drop leakage-prone columns
leakage_columns = [col for col in df.columns if col.lower() in ["is deal closed", "is closed won", "is closed lost"]]
df.drop(columns=leakage_columns, inplace=True, errors='ignore')

# Drop all date-related columns (date/time/duration fields)
date_cols_to_drop = [
    col for col in df.columns
    if 'date' in col.lower() or 'time' in col.lower() or 'contract start' in col.lower() or 'contract end' in col.lower()
]
df.drop(columns=date_cols_to_drop, inplace=True, errors='ignore')

# Define features and target
non_feature_columns = [
    'Target_Won', 'Deal Name', 'Deal Description', 'Record ID_x', 'Record ID_y',
    'Associated Company_x', 'Associated Company_y', 'Associated Company (Primary)', 'Company name'
]
X = df.drop(columns=non_feature_columns, errors='ignore')
y = df['Target_Won']
```

```python
# Cell 2 (updated): Numerical feature importance (graph + top 7 safe features)

# Identify numeric-only columns (after dropping dates)
original_numerical = X.select_dtypes(include=[np.number]).columns.tolist()
X_numerical_full = X[original_numerical].fillna(-999)

# Train/test split and model
X_train_num, X_test_num, y_train_num, y_test_num = train_test_split(X_numerical_full, y, test_size=0.2, random_state=42)
clf_num = RandomForestClassifier(n_estimators=100, random_state=42)
clf_num.fit(X_train_num, y_train_num)

# Feature importance calculation
importances_num = clf_num.feature_importances_
feat_importances_num = pd.Series(importances_num, index=X_numerical_full.columns).sort_values(ascending=False)

# Plot top 20 numerical features
plt.figure(figsize=(12, 8))
sns.barplot(x=feat_importances_num[:20], y=feat_importances_num.index[:20])
plt.title('Top 20 Numerical Feature Importances (No Date Columns)')
plt.xlabel('Importance Score')
plt.ylabel('Numerical Feature')
plt.tight_layout()
plt.show()

# Select top 7 safe numerical features
safe_numerical_features = [
    col for col in feat_importances_num.index
    if all(kw not in col.lower() for kw in ["probability", "weighted", "forecast", "close", "is closed", "deal probability"])
][:7]

X_numerical = X[safe_numerical_features].copy()
for col in X_numerical.columns:
    X_numerical[col].fillna(X_numerical[col].median(), inplace=True)
```

```python
# Cell 3: Categorical feature encoding, importance graph, and top 3 selection

# Identify categorical columns
original_categorical = X.select_dtypes(include=['object']).columns.tolist()

# Label encode all safe categorical columns (excluding "Deal Stage")
label_encoded_df = pd.DataFrame(index=X.index)
for col in original_categorical:
    if 'deal stage' in col.lower():
        continue
    le = LabelEncoder()
    label_encoded_df[col] = le.fit_transform(X[col].astype(str))

# Handle missing values
label_encoded_df.fillna(-1, inplace=True)

# Train/test split and model
X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(label_encoded_df, y, test_size=0.2, random_state=42)
clf_cat = RandomForestClassifier(n_estimators=100, random_state=42)
clf_cat.fit(X_train_cat, y_train_cat)

# Feature importance
feat_importances_cat = pd.Series(clf_cat.feature_importances_, index=label_encoded_df.columns).sort_values(ascending=False)

# Plot top 20 categorical features
plt.figure(figsize=(12, 8))
sns.barplot(x=feat_importances_cat[:20], y=feat_importances_cat.index[:20])
plt.title('Top 20 Categorical Feature Importances (No Date Columns)')
plt.xlabel('Importance Score')
plt.ylabel('Categorical Feature')
plt.tight_layout()
plt.show()

# Select top 3 safe categorical features
safe_categorical_features = [
    col for col in feat_importances_cat.index
    if all(kw not in col.lower() for kw in ['forecast', 'close', 'won', 'lost', 'deal stage'])
][:3]
X_categorical = label_encoded_df[safe_categorical_features].copy()
for col in X_categorical.columns:
    X_categorical[col].fillna(X_categorical[col].mode()[0], inplace=True)
```

```python
# Cell 4: Train and evaluate final baseline model using 10 features (7 numerical + 3 categorical)

# Combine selected features
X_selected = pd.concat([X_numerical, X_categorical], axis=1)
```
