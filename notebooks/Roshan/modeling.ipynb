```python
# Convert Create Date to datetime
companies_df['Create Date'] = pd.to_datetime(companies_df['Create Date'], errors='coerce')

# Create date-based fields
companies_df['Create_Year'] = companies_df['Create Date'].dt.year
companies_df['Create_Month'] = companies_df['Create Date'].dt.month
companies_df['Create_Quarter'] = companies_df['Create Date'].dt.quarter
companies_df['Create_YearMonth'] = companies_df['Create Date'].dt.to_period('M').astype(str)

# 1. Standardized Industry (simple title-case normalization)
companies_df['Industry_Standardized'] = companies_df['Industry'].str.title().str.strip()

# 2. Company Size Category (based on 'Total Agents')
def company_size_category(agents):
    if pd.isnull(agents):
        return 'Unknown'
    elif agents < 10:
        return 'Very Small'
    elif agents < 50:
        return 'Small'
    elif agents < 250:
        return 'Medium'
    elif agents < 1000:
        return 'Large'
    else:
        return 'Enterprise'

companies_df['Company_Size_Category'] = companies_df['Total Agents'].apply(company_size_category)

# 3. Revenue Category
def revenue_category(revenue):
    if pd.isnull(revenue):
        return 'Unknown'
    elif revenue < 1e6:
        return '<$1M'
    elif revenue < 1e7:
        return '$1M-$10M'
    elif revenue < 5e7:
        return '$10M-$50M'
    elif revenue < 1e8:
        return '$50M-$100M'
    else:
        return '>$100M'

companies_df['Revenue_Category'] = companies_df['Annual Revenue'].apply(revenue_category)

# 4. Region mapping (simplified)
def map_region(country):
    if pd.isnull(country):
        return 'Unknown'
    country = country.strip().lower()
    if country in ['united states', 'canada', 'mexico']:
        return 'North America'
    elif country in ['brazil', 'argentina', 'chile']:
        return 'South America'
    elif country in ['united kingdom', 'germany', 'france', 'netherlands']:
        return 'Europe'
    elif country in ['india', 'china', 'japan', 'singapore']:
        return 'Asia'
    elif country in ['australia', 'new zealand']:
        return 'Oceania'
    else:
        return 'Other'

companies_df['Region'] = companies_df['Country/Region'].apply(map_region)

# 5. Technology_Count from "Web Technologies"
companies_df['Technology_Count'] = companies_df['Web Technologies'].fillna('').apply(lambda x: len(x.split(';')) if x else 0)

# 6. Uses_[Technology] flags (example: uses Google Tag Manager)
companies_df['Uses_Google_Tag_Manager'] = companies_df['Web Technologies'].fillna('').str.contains('Google Tag Manager', case=False).astype(int)

# 7. Is_BPO flag
companies_df['Is_BPO'] = companies_df['BPO Program'].notna().astype(int)
```

## Choosing the features

```python
# Drop obviously irrelevant or identifier columns
columns_to_exclude = [
    'Target_Won', 'Is Deal Closed?', 'Is Closed Won', 'Is closed lost',
    'Record ID_x', 'Record ID_y', 'comp_id', 'Deal Name', 'Company name',
    'Associated Company_y', 'Associated Company (Primary)'
]

# Create a working copy of the dataset
features_df = binary_df.drop(columns=columns_to_exclude, errors='ignore')

# Check numeric correlation with the target
correlations = binary_df.corr(numeric_only=True)['Target_Won'].drop('Target_Won').sort_values(ascending=False)

# Get top correlated numeric features (threshold > 0.1 absolute correlation)
useful_numeric_features = correlations[abs(correlations) > 0.1].index.tolist()

# Also include some categorical columns that are likely useful based on domain knowledge
likely_useful_categoricals = [
    'Deal Stage', 'Deal Type', 'Industry_Standardized', 'Company_Size_Category',
    'Revenue_Category', 'Country/Region', 'Forecast category', 'Pipeline',
    'Original Traffic Source', 'Deal source attribution 2'
]

# Filter for features that are actually in the dataframe
likely_useful_categoricals = [col for col in likely_useful_categoricals if col in features_df.columns]

# Combine selected features
selected_features = useful_numeric_features + likely_useful_categoricals + ['Target_Won']

selected_features
```

## Now, for the second one, the target is gonna be multi-class

```python
# Define a function to create the multi-class target column
def get_deal_status(row):
    if not row['Is Deal Closed?']:
        return 'Open'
    elif row['Is Closed Won']:
        return 'Won'
    elif row['Is closed lost']:
        return 'Lost'
    return 'Unknown'

# Apply the function to create the new column
cleaned_df['Deal_Status'] = cleaned_df.apply(get_deal_status, axis=1)

# Show value counts for sanity check
cleaned_df['Deal_Status'].value_counts()
```

```python
# Select features and target
selected_columns = [
    'Deal probability', 'Weighted amount',
    'Weighted amount in company currency', 'Number of Sessions',
    'Number of Form Submissions', 'Is_BPO', 'Number of Pageviews',
    'Deal Type', 'Industry_Standardized', 'Company_Size_Category',
    'Revenue_Category', 'Country/Region', 'Forecast category', 'Pipeline',
    'Original Traffic Source', 'Deal source attribution 2', 'Deal_Status'
]

model_df = cleaned_df[selected_columns]
```

```python
# Calculate missing value percentage
missing_percentage = (model_df.isnull().mean() * 100).sort_values(ascending=False)

missing_percentage
```

```python
# Identify numeric and categorical columns
numeric_cols = model_df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = [col for col in model_df.columns if col not in numeric_cols + ['Deal_Status']]

# Impute numeric with median
for col in numeric_cols:
    model_df[col].fillna(model_df[col].median(), inplace=True)

# Impute categorical with mode
for col in categorical_cols:
    model_df[col].fillna(model_df[col].mode()[0], inplace=True)

# Recheck to confirm all missing values are handled
model_df.isnull().sum().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Separate features and target
X = model_df.drop(columns=["Deal_Status"])
y = model_df["Deal_Status"]

# Identify categorical columns
categorical_features = X.select_dtypes(include=["object", "category"]).columns.tolist()

# Apply ordinal encoding
encoder = OrdinalEncoder()
X[categorical_features] = encoder.fit_transform(X[categorical_features])

# Output shape and preview
print(X.shape)
X.head()
```
