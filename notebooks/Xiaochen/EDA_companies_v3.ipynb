```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Cleaning
file_path = '/content/drive/MyDrive/Colab_Notebooks/project/anonymized_hubspot_companies.csv'
df = pd.read_csv(file_path, low_memory=False)
```

```python
print("DataFrame Info:")
print(df.info())

print("\nMissing Values per Column:")
print(df.isnull().sum())

duplicate_count = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicate_count)

print("\nFirst 5 Rows:")
print(df.head())
```

```python
threshold = 0.8
df = df.loc[:, df.isnull().mean() < threshold].copy()

zero_threshold = 0.8
num_columns = df.select_dtypes(include=[np.number]).columns
cols_to_drop = [col for col in num_columns if (df[col] == 0).mean() > zero_threshold]
df = df.drop(columns=cols_to_drop)

num_columns = df.select_dtypes(include=[np.number]).columns

cat_columns = df.select_dtypes(include=['object', 'category']).columns
for col in cat_columns:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

for col in num_columns:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)

print("DataFrame shape after cleaning:", df.shape)
print(df.info())
```

This block performs data cleaning by removing columns with over 80% missing or zero values. It then fills missing values using the mode for categorical features and the median for numeric ones

```python
# Date Conversion
print(df[['Create Date', 'Last Modified Date']].head())

df['Create Date'] = pd.to_datetime(df['Create Date'], format='%Y-%m-%d %H:%M', errors='coerce')
df['Last Modified Date'] = pd.to_datetime(df['Last Modified Date'], format='%Y-%m-%d %H:%M', errors='coerce')

print(df[['Create Date', 'Last Modified Date']].head())
print(df.info())
```

```python
# Feature Engineering
df['duration'] = (df['Last Modified Date'] - df['Create Date']).dt.days

print("\n'duration' feature sample values:")
print(df['duration'].head())
print(df.info())
```

This step creates a new feature duration by calculating the number of days between Create Date and Last Modified Date

```python
num_columns = df.select_dtypes(include=[np.number]).columns
# Outlier Handling
for col in num_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print(df.shape)
print(df.columns)
```

This step handles outliers in numeric columns by applying the IQR method and capping extreme values at the lower and upper bounds

```python
df.drop(columns=["Record ID"], inplace=True)
num_columns = df.select_dtypes(include=[np.number]).columns
```

This step removes the "Record ID" column, which likely doesnâ€™t carry analytical value, and then updates the list of numeric columns for further analysis

```python
# Summary Statistics
numeric_stats = df[num_columns].describe().T
numeric_stats.rename(columns={'50%': 'median'}, inplace=True)

print("\nSummary Statistics (Mean, Median, Standard Deviation, Min, Max) for Numeric Columns:")
print(numeric_stats[['mean', 'median', 'std', 'min', 'max']])
```

```python
# Class distribution
for col in cat_columns:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts())
```

```python
# correlation matrix
corr_matrix = df[num_columns].corr()
print("\nCorrelation Matrix:")
print(corr_matrix)
```

This step calculates the correlation matrix for all numeric features to understand how they relate to each other. A higher absolute value indicates a stronger linear relationship

Insight:
1. Annual Revenue has a moderate positive correlation with Number of Employees (0.50), meaning larger companies tend to have higher revenue.
2. duration is moderately correlated with Number of times contacted (0.36), indicating that longer-lasting records were likely contacted more.

This step plots histograms for each numeric column to visualize their distributions

What the results show:

5. Duration: Shows clustering around certain values, especially around 400â€“450 days, possibly reflecting a common data collection behavior

```python
# Top 10 Class Distribution with Percentages for Categorical Features
for col in cat_columns:
    print(f"\nColumn: {col}")
    counts_percent = df[col].value_counts(normalize=True).head(10) * 100
    print(counts_percent)
    counts_percent.iloc[::-1].plot(kind='barh', title=f"Top 10 Class Distribution (%) for {col}")
    plt.xlabel("Percentage (%)")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

This step summarizes the top 10 most frequent categories in each categorical variable and calculates their percentage shares, helping us understand the dominant classes and detect potential class imbalance within each feature

The results reveal that many fields exhibit strong category concentration â€” for example, over 80% of records fall under â€œTier 2: Partial Fitâ€ in ICP Fit Level; Primary Industry and Primary Sub-Industry are heavily dominated by Manufacturing and Banking, respectively; nearly 78% of entries in Country/Region are from the United States, suggesting the dataset is largely U.S.-centric. The Last Modified Date is concentrated on a few specific timestamps, possibly due to batch updates or system imports. Some fields, such as Associated Contact, also show a high level of repetition. These findings help us understand the structure of the data, identify skewed distributions, and provide useful context for further analysis

```python
df['Create_Month'] = df['Create Date'].dt.to_period('M')

# time-series trends: Count the number of companies created per month
company_count = df.groupby('Create_Month').size()
company_count.index = company_count.index.to_timestamp()

plt.figure(figsize=(10, 6))
plt.plot(company_count.index, company_count.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Number of Companies Created")
plt.title("Trend of Companies Created Over Time")
plt.grid(True)
plt.show()
```

This step extracts the month from each companyâ€™s creation date, counts how many companies were created each month, and plots the trend over time

What the result shows:

The chart reveals a significant spike in October 2023. Aside from this peak, the number of companies created varies across months without a strong seasonal pattern, though higher activity is also observed in early 2023 and mid-2024

```python
#  time-series trends: Aggregate 'Annual Revenue' by month
revenue_monthly = df.groupby('Create_Month')['Annual Revenue'].sum()
revenue_monthly.index = revenue_monthly.index.to_timestamp()

plt.figure(figsize=(10, 6))
plt.plot(revenue_monthly.index, revenue_monthly.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Total Annual Revenue")
plt.title("Trend of Total Annual Revenue Over Time")
plt.grid(True)
plt.show()
```

This step aggregates the total Annual Revenue of newly created companies for each month and visualizes the trend over time

What the result shows:

There is a sharp spike in October 2023, indicating that a small number of companies with extremely high revenue were likely added that month. Outside this outlier, most months show lower and fluctuating total revenues, without a strong seasonal trend.
